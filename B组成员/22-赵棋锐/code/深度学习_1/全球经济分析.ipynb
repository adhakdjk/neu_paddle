```python
import pandas as pd
import numpy as np
# 在work 文件夹中的文件Countries GDP 1960-2020.csv 中储存了一些国家从1960-2020年的GDP

data = pd.read_csv('/home/aistudio/work/Countries GDP 1960-2020.csv')
data
```

```python
#请写出1960年GDP最高的国家

ceshi_data=data['1960'].dropna()
mean_1960=ceshi_data.mean()
data['1960']=data['1960'].fillna(mean_1960)
data=data.drop(data[data['Country Name'].isin(['World'])].index[0])
data=data.drop(data[data['Country Name'].isin(['High income'])].index[0])
data=data.drop(data[data['Country Name'].isin(['OECD members'])].index[0])
data=data.drop(data[data['Country Name'].isin(['Post-demographic dividend'])].index[0])
data=data.drop(data[data['Country Name'].isin(['North America'])].index[0])
#data=data.drop(data[data['Country Name'].isin(['East Asia & Pacific (IDA & IBRD countries)','East Asia & Pacific (excluding high income)','Lower middle income','East Asia & Pacific','Early-demographic dividend','Late-demographic dividend','Euro area','Upper middle income','IBRD only','Middle income','Low & middle income','IDA & IBRD total'])].index[0])
data=data.drop(data[data['Country Name'].isin(['Euro area'])].index[0])
data=data.drop(data[data['Country Name'].isin(['IDA & IBRD total'])].index[0])
data=data.drop(data[data['Country Name'].isin(['Low & middle income'])].index[0])
data=data.drop(data[data['Country Name'].isin(['Middle income'])].index[0])
data=data.drop(data[data['Country Name'].isin(['IBRD only'])].index[0])
data=data.drop(data[data['Country Name'].isin(['Upper middle income'])].index[0])
data=data.drop(data[data['Country Name'].isin(['East Asia & Pacific'])].index[0])
data=data.drop(data[data['Country Name'].isin(['Late-demographic dividend'])].index[0])
data=data.drop(data[data['Country Name'].isin(['East Asia & Pacific (excluding high income)'])].index[0])
data=data.drop(data[data['Country Name'].isin(['Early-demographic dividend'])].index[0])
data=data.drop(data[data['Country Name'].isin(['East Asia & Pacific (IDA & IBRD countries)'])].index[0])

data=data.drop(data[data['Country Name'].isin(['Lower middle income'])].index[0])
data=data.drop(data[data['Country Name'].isin(['Latin America & the Caribbean (IDA & IBRD countries)'])].index[0])
```
